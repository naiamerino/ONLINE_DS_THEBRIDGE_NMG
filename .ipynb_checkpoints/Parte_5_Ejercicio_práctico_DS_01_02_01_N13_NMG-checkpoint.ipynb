## Ejercicio 2
Haz un programa que detecte todos los duplicados de un elemento en una lista:
* Utiliza una variable duplicado para asignarle el valor del elemento del que queremos encontrar todos sus duplicados.
* El programa debe devolver todos los índices de los elementos duplicados.
* Aplícalo para encontrar los duplicados de "un", "es" y "binario" de:  

>```ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]```


`````python
def devuelve_duplicados (lista, duplicado):
    indices = []
    if (lista.count (duplicado) > 1):
        for i in range(len(lista)):
            if lista[i] == duplicado:
                indices.append (i)                   
    return (indices)

ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]
resultado = devuelve_duplicados (ejer_3, "un")
print(resultado)
resultado = devuelve_duplicados (ejer_3, "es")
print (resultado)
resultado = devuelve_duplicados (ejer_3, "binario")
print (resultado)

    
    
`````

## Ejercicio 12
Solicitar al usuario que ingrese su dirección email. Imprimir un mensaje indicando si la dirección es válida o no, valiéndose de una función para decidirlo. Una dirección se considerará válida si contiene el símbolo "@".

`````python

def es_la_direccion_valida (email):
    if "@" in set(email):
        print ("Direccion de email valida")
    else:
        print ("Direccion de email no valida")

dir_email = input ("Introduzca su dirección de email")
es_la_direccion_valida (dir_email)
`````
