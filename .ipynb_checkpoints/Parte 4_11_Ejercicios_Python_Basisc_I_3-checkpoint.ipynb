## Ejercicio 1
Tenemos las siguientes variables:

```Python
A = 4
B = "Text"
C = 4.1
```

Comprueba:
1. Si A y B son equivalentes
2. Si A y C NO son equivalentes
3. Si A es mayor que C
4. Si C es menor o igual que A
5. Si B NO es equivalente a C

````python
A = 4
B = "Text"
C = 4.1

print (A == B,"\n", A != C,"\n", A > C,"\n", C <= A,"\n", B != C)
````

## Ejercicio 2
Crea un programa donde se recojan dos inputs del usuario, y el output del programa sea si esos inputs son iguales o no

````python
input_1 =  input ("Pon algo: ")
input_2 =  input ("Pon algo (igual o diferente de lo anterior):")
if (input_1 == input_2):
    print ("Las dos entradas son iguales")
else: 
    print ("Las dos entradas son diferentes")

                  
````

## Ejercicio 3
Mismo programa que en 2, pero en esta ocasión tienen que ser tres inputs y dos salidas. Una de las salidas que nos indique si todos son iguales, y la otra si al menos dos inputs sí que lo son

````python
input_1 =  input ("Pon algo: ")
input_2 =  input ("Pon algo (igual o diferente de lo anterior):")
input_3 = input ("Pon una tercera cosa también igual o diferente a las anteriores")

print ("Todas son iguales:", input_1 == input_2 and input_2 == input_3)
print ("Al menos dos son iguales:", input_1 == input_2 or input_2 == input_3 or input_1 == input_3)

````
